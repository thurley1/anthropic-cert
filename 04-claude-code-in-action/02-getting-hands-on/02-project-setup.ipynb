# Project Setup

Getting Claude oriented to a codebase so it follows your conventions.


## `/init` and `CLAUDE.md`

Run **`/init`** in a project → Claude scans the codebase and writes a **`CLAUDE.md`** capturing structure, build/test commands, conventions, and patterns. This file is **auto-loaded as context** every session, so Claude "remembers" the project.

**Scopes (all merged into context):**

| File | Scope | In git? |
|------|-------|---------|
| `./CLAUDE.md` | project (shared) | yes |
| `./CLAUDE.local.md` | your personal project notes | no (gitignored) |
| `~/.claude/CLAUDE.md` | user (all your projects) | n/a |

Nested `CLAUDE.md` files in subdirectories layer on when Claude works in that subtree.

## Adding to memory with `#`

Type **`#`** followed by a rule (e.g. `# always run the tests before committing`) and Claude offers to save it to project / local / user memory — a fast way to grow `CLAUDE.md` as you notice conventions.

**Good `CLAUDE.md` content:** build/lint/test commands, code style, architectural rules, "gotchas," and anything you find yourself re-explaining. Keep it tight — it's spent on every request.

## Multiple directories

`--add-dir <path>` at launch (or `/add-dir` in-session) lets Claude work across more than one folder (e.g. a frontend + backend repo).
