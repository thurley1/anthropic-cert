# Controlling Context

A long session fills the context window with old files, tool output, and tangents. Managing it keeps Claude sharp and costs down.


## The key commands

- **`/clear`** — wipe the conversation and start fresh. Do this **between unrelated tasks** so stale context doesn't bleed in (and mislead the model).
- **`/compact`** — summarize the conversation so far into a compact form, freeing space while keeping the gist. Happens **automatically** as you approach the window limit; you can also trigger it manually (optionally with a focus hint, e.g. `/compact keep the auth refactor details`).
- **`/context`** — inspect what's currently using the context window (files, history, tool results, system prompt/CLAUDE.md).

## Principles

- **Right-size the context** — more relevant context helps; irrelevant context hurts (distraction + cost). Be selective about what you `@`-mention.
- **`CLAUDE.md` is persistent context** — put durable project rules there so you don't re-explain them every session (and they survive `/clear`).
- **Reset early and often** — a fresh `/clear` before a new task usually beats dragging a huge history along.
- **Long autonomous runs** rely on auto-compaction to keep going past a single window.
