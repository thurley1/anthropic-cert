# Making Changes

How Claude Code actually edits code, and the workflow that gets reliable results.


## The tools that change things

- **Edit** — targeted find/replace edits to a file (shows you a diff).
- **Write** — create or overwrite a file.
- **Bash** — run commands (tests, builds, installers, git).

Claude proposes edits as **diffs**; depending on your permission mode you approve them or they auto-apply. Review diffs like you would a teammate's PR.

## The reliable workflow: context → plan → implement

1. **Context** — point Claude at the relevant files (previous lesson).
2. **Plan first** — ask for an approach/steps and *explicitly say "don't write code yet."* Review it. (Claude Code has a **Plan Mode** — cycle input modes with `Shift+Tab` — that researches read-only and proposes a plan before touching anything.)
3. **Implement** — once the plan's good, let it build, run tests, and iterate.

**TDD variant:** have Claude brainstorm tests → write the tests → write code until they pass. Clear success criteria → more robust output.

## Let it drive the chores

Claude can stage and commit to git, run the test suite, manage dependencies, and fix failures — ask it instead of switching to the terminal. (It writes reasonable commit messages; review before pushing.)

> Permission modes control how much it does without asking (from "ask each time" up to more autonomous modes) — pick what you're comfortable with.
