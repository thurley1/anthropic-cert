# Gotchas Around Hooks

After running `npm run setup` you'll notice **two** settings files in `.claude/` — `settings.example.json` and a generated `settings.local.json`. Here's why.

## Hooks security: use absolute paths

Claude Code's docs recommend security practices for hooks. A key one: **use absolute paths** for hook scripts (not relative paths). This mitigates **path-interception** and **binary-planting** attacks — a relative path could resolve to a malicious script placed earlier on the path.

## The sharing problem

Absolute paths make `settings.json` **hard to share**: the absolute path to a hook script on your machine differs from anyone else's (the project lives in a different directory per person). A committed settings file with a hard-coded absolute path breaks for everyone else.

## The fix: `$PWD` placeholder + `init-claude.js`

The project ships a **`settings.example.json`** whose script references use a **`$PWD` placeholder** instead of a real absolute path.

Running **`npm run setup`**:
1. installs dependencies, and
2. runs **`scripts/init-claude.js`**, which **replaces `$PWD`** with the absolute path to the project on *your* machine, copies the file, and renames it to **`settings.local.json`**.

So you get the best of both: a **shareable** example file in git, and a generated local file that uses the **recommended absolute paths**.

> That's why there are two files: `settings.example.json` (committed, `$PWD` placeholder) and `settings.local.json` (generated, machine-specific, gitignored). In the `queries` project, `scripts/init-claude.js` is the generator.
