# The Claude Code SDK (Agent SDK)

> **Erratum in the course:** the *video* uses an older package name that no longer works. The current package is **`@anthropic-ai/claude-agent-sdk`** — follow that.

The **Agent SDK** runs Claude Code **programmatically** from your own scripts/apps. Available for **TypeScript and Python**, it gives you the same agent loop the CLI uses — file reading, editing, tool use — under your control. (This is the mechanism the query-duplication hook used to launch a reviewing Claude.)

## Install

```bash
mkdir sdk-demo
cd sdk-demo
npm init -y
npm install @anthropic-ai/claude-agent-sdk
```

⚠️ The package is **`@anthropic-ai/claude-agent-sdk`**. The similarly-named **`@anthropic-ai/claude-code`** is the **CLI itself** and can't be imported. (Python users: `pip install claude-agent-sdk`.)

## Minimal example

`index.mjs`:
```javascript
import { query } from "@anthropic-ai/claude-agent-sdk";

const prompt = "List the files in the current directory";

for await (const message of query({ prompt })) {
  console.log(JSON.stringify(message, null, 2));
}
```

Run it:
```bash
node index.mjs
```

You'll get a **stream of JSON messages** — the same conversation events as the CLI: tool calls, tool results, and Claude's text. `query()` returns an async iterable you loop over with `for await`.

## Restricting tools

By default the SDK has the full tool set. Narrow it with **`allowedTools`**:

```javascript
for await (const message of query({
  prompt,
  options: { allowedTools: ["Read", "Glob"] },
})) {
  // ...
}
```

This is the SDK equivalent of the CLI's `--allowedTools` flag — exactly how you'd sandbox a programmatic agent (e.g. a review-only hook that may read but not write).

## Where to go next

The SDK supports everything the CLI does: **custom system prompts, MCP servers, hooks, subagents, session resumption**. See the Agent SDK docs for the full reference.

> **Disambiguation** (worth keeping straight): the **Agent SDK** (`@anthropic-ai/claude-agent-sdk`) is the full Claude Code harness as a library — built-in tools, the loop, hooks, subagents. It's *different* from the plain Anthropic API SDK's **Tool Runner** (`client.beta.messages.tool_runner`), which just loops over tools *you* define with no built-in filesystem tools. Both are "harness, you host"; the Agent SDK is the batteries-included coding agent.
