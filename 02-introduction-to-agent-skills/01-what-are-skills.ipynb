# What Are Skills?

Every time you re-explain your team's coding standards, PR-review format, or commit-message style to Claude, you're repeating yourself. **Skills** fix that: write the instructions **once**, and Claude Code applies them **automatically** whenever they're relevant.

## What skills are

A skill is a **folder of instructions/resources** Claude Code can discover and use to handle a task more accurately. Each skill lives in a **`SKILL.md`** file with a **`name`** and **`description`** in its frontmatter:

```markdown
---
name: pr-review
description: Reviews pull requests for code quality. Use when reviewing PRs or checking code changes.
---
Below the frontmatter you write the actual instructions — your review checklist,
formatting preferences, or whatever Claude needs for that task.
```

**The `description` is how Claude decides to use the skill.** When you make a request, Claude compares it against all available skill descriptions and activates the matching one(s).

## Where skills live

| Location | Scope |
|----------|-------|
| **`~/.claude/skills`** (Windows: `C:/Users/<user>/.claude/skills`) | **Personal** — follow you across all projects (your commit style, doc format, how you like code explained) |
| **`.claude/skills`** in a repo root | **Project** — committed to version control, so anyone who clones the repo gets them (team standards, brand guidelines) |

## Skills vs. CLAUDE.md vs. slash commands

What makes skills unique: they're **automatic** and **task-specific**.

| Mechanism | When it loads |
|-----------|---------------|
| **CLAUDE.md** | into **every** conversation (e.g. "always use TypeScript strict mode") |
| **Skills** | **on demand**, when your request matches the skill's description. Only the name+description are loaded initially, so they don't fill the context window — the full checklist loads only when you actually ask for that task |
| **Slash commands** | only when you **explicitly type** them |

So your PR-review checklist isn't in context while you debug — it loads when you ask for a review. When Claude matches a skill, you see it load in the terminal.

## When to use skills

Best for **specialized knowledge tied to specific tasks**:
- Code-review standards your team follows
- Commit-message formats you prefer
- Brand guidelines for your org
- Documentation templates
- Debugging checklists for particular frameworks

**Rule of thumb:** if you find yourself explaining the same thing to Claude repeatedly, that's a skill waiting to be written.
