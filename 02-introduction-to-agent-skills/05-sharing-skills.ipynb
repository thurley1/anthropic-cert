# Sharing Skills

A skill only you use is helpful; the same skill shared across a team **standardizes** the work. Three distribution methods, plus one important gotcha about subagents.

## 1. Commit to your repository (simplest)

Put skills in **`.claude/skills`** and commit them — anyone who clones the repo gets them automatically, no install. Push updates → everyone gets them on the next pull.

Best for: team coding standards, project-specific workflows, skills that reference your codebase structure. (The `.claude/` dir holds agents, hooks, skills, and settings — all version-controlled and shared via normal Git.)

## 2. Distribute through plugins

Plugins extend Claude Code and are built to be **shared across teams/projects**. In a plugin project, add a `skills/` directory structured like `.claude/skills` (each skill = a folder with a `SKILL.md`). Publish the plugin to a **marketplace**, and others discover + install it.

Best when the skills **aren't too project-specific** and are useful beyond your immediate team.

## 3. Enterprise managed settings

Admins can deploy skills **organization-wide** via managed settings. **Enterprise skills take the highest priority** — they override personal, project, and plugin skills of the same name.

Managed settings support controls like `strictKnownMarketplaces` (where plugins may be installed from):

```json
"strictKnownMarketplaces": [
  { "source": "github", "repo": "acme-corp/approved-plugins" },
  { "source": "npm", "package": "@acme-corp/compliance-plugins" }
]
```

Right for **mandatory** standards — security requirements, compliance workflows, practices that *must* be consistent org-wide.

## Skills & subagents (the gotcha)

**Subagents don't automatically see your skills** — a delegated subagent starts with a fresh, clean context. Key distinctions:

- **Built-in agents** (Explorer, Plan, Verify) **can't access skills at all**.
- **Custom subagents** you define **can** use skills — but only when you **explicitly list** them.
- Skills are loaded **when the subagent starts**, not on demand like in the main conversation.

Create a custom subagent (an agent markdown file in **`.claude/agents`**, or via **`/agents`**) with a **`skills`** field in its frontmatter:

```markdown
---
name: frontend-security-accessibility-reviewer
description: "Use this agent when you need to review frontend code for accessibility..."
tools: Bash, Glob, Grep, Read, WebFetch, WebSearch, Skill...
model: sonnet
color: blue
skills: accessibility-audit, performance-check
---
```

The skills must already exist in `.claude/skills`. Then that subagent loads both skills and applies them to every review.

**Great when:** isolated task delegation with specific expertise, different subagents needing different skills (frontend vs. backend reviewer), and enforcing standards in delegated work without relying on prompts.
