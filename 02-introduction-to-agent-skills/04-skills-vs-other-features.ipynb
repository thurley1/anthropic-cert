# Skills vs. Other Claude Code Features

Claude Code has several customization options — **skills, CLAUDE.md, subagents, hooks, MCP servers**. They solve *different* problems; picking the wrong one adds complexity. The fix is knowing each one's specialty and **combining** them.

## At a glance

| Feature | Trigger / nature | Best for |
|---------|------------------|----------|
| **CLAUDE.md** | loads into **every** conversation | always-on project standards |
| **Skills** | **request-driven**, load **on demand** when matched | task-specific expertise |
| **Subagents** | run in an **isolated** context, return results | delegated work needing isolation / different tools |
| **Hooks** | **event-driven** (file save, tool call) | automated side effects / validation |
| **MCP servers** | external **tools & integrations** | connecting outside services (different category) |

## CLAUDE.md vs Skills

- **CLAUDE.md** — always in context. TypeScript strict mode, "never modify the DB schema," framework/style preferences.
- **Skills** — join the conversation only when matched. Your PR-review checklist isn't in context while writing new code; it activates when you ask for a review.

**CLAUDE.md** for: project-wide always-apply standards, hard constraints, style.
**Skills** for: task-specific expertise, sometimes-relevant knowledge, detailed procedures that would clutter every conversation.

## Skills vs Subagents

- **Skills** add knowledge to your **current** conversation (instructions join existing context).
- **Subagents** run in a **separate** context — receive a task, work independently, return results.

**Subagents** when: you want to delegate to an isolated context, need different tool access, or want isolation between delegated work and your main thread.
**Skills** when: you want to enhance Claude's knowledge for the current task, and the expertise applies throughout the conversation.

## Skills vs Hooks

- **Hooks** fire on **events** (run a linter on every file save, validate before a tool call). Event-driven.
- **Skills** activate on **what you're asking**. Request-driven.

**Hooks** for: operations on every save, validation before specific tools, automated side effects of Claude's actions.
**Skills** for: knowledge that informs *how* Claude handles requests; guidelines that affect its reasoning.

## Putting it all together

A typical setup uses several at once, each in its lane:

- **CLAUDE.md** — always-on project standards
- **Skills** — task-specific expertise, on demand
- **Hooks** — automated operations triggered by events
- **Subagents** — isolated execution for delegated work
- **MCP servers** — external tools & integrations

Don't force everything into skills when another option fits better — **combine** them. Reach for a **skill** when you have knowledge Claude should apply **automatically when the topic is relevant**.
