# Troubleshooting Skills

When a skill misbehaves, the problem almost always falls into one of a few predictable buckets: it **doesn't trigger**, **doesn't load**, **conflicts** with another skill, or **fails at runtime**. Most fixes are straightforward once you know the bucket.

## Start with the skills validator

Before debugging anything else, run the **agent skills verifier** — it catches structural problems up front. Installation varies by OS; **`uv`** is the quickest way to get it.

Once installed, `cd` into your skill directory (or run it from anywhere) and let it check the structure. This alone resolves a surprising number of "why isn't it working?" cases.

## Skill doesn't trigger

Exists, passes validation, but Claude won't use it when you expect. **The cause is almost always the description.**

Claude uses **semantic matching** — your request must overlap with the description's meaning. Not enough overlap → no match. Fix it:

1. Compare the description against **how you actually phrase requests**.
2. Add the trigger phrases you'd really say.
3. Test variations: *"help me profile this," "why is this slow?", "make this faster."*
4. If a variation fails to trigger, add those keywords to the description.

## Skill doesn't load

Doesn't even appear when you ask *"what skills are available?"* — check the **structural requirements**:

- `SKILL.md` must be inside a **named directory**, not at the skills root.
- The filename must be **exactly `SKILL.md`** — all-caps `SKILL`, lowercase `md`.

Run **`claude --debug`** to see loading errors; look for messages mentioning your skill name. Often that alone points straight at the problem.

## Wrong skill gets used

If Claude picks the wrong skill or seems confused between two, your **descriptions are too similar**. Make them **distinct**. Being specific doesn't just help Claude decide *when* to use a skill — it also prevents collisions with similar-sounding ones (`frontend-review` / `backend-review`, not two vague `review`s).

## Priority conflicts (being shadowed)

If your **personal** skill is ignored, a **higher-priority** skill of the same name may be winning. Recall the hierarchy: **Enterprise → Personal → Project → Plugins**.

Example: an enterprise `code-review` and your personal `code-review` → the **enterprise one wins every time**. Options:

- **Rename** your skill to something more distinct (usually the easier path).
- Talk to your **admin** about the enterprise skill.

## Plugin skills not appearing

Installed a plugin but its skills don't show up?

1. **Clear the cache**, restart Claude Code, and **reinstall**.
2. If they still don't appear, the **plugin structure** may be wrong — this is where the **validator** earns its keep.

## Runtime errors

The skill loads but fails during execution. Common causes:

- **Missing dependencies** — external packages must be installed. Add dependency info to the skill **description** so Claude knows what's needed.
- **Permission issues** — scripts need execute permission: **`chmod +x`** on any scripts the skill references.
- **Path separators** — use **forward slashes everywhere**, even on Windows.

## Quick troubleshooting checklist

| Symptom | Fix |
|---------|-----|
| **Not triggering** | Improve the description; add trigger phrases |
| **Not loading** | Check path, exact `SKILL.md` filename, YAML syntax |
| **Wrong skill used** | Make descriptions more distinct from each other |
| **Being shadowed** | Check the priority hierarchy; rename if needed |
| **Plugin skills missing** | Clear cache and reinstall |
| **Runtime failure** | Check dependencies, permissions, and paths |
