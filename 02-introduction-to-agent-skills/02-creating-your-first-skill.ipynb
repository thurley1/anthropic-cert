# Creating Your First Skill

We build a **personal** `pr-description` skill that writes PR descriptions in a consistent format. Personal → lives in your home dir → works across all projects.

## 1. Make the skill directory

The directory name should match the skill name:

```bash
mkdir -p ~/.claude/skills/pr-description
```
(Windows: `C:/Users/<you>/.claude/skills/pr-description`)

## 2. Create `SKILL.md`

Inside that directory, a `SKILL.md` with frontmatter (metadata) + instructions:

```markdown
---
name: pr-description
description: Writes pull request descriptions. Use when creating a PR, writing a PR, or when the user asks to summarize changes for a pull request.
---

When writing a PR description:

1. Run `git diff main...HEAD` to see all changes on this branch
2. Write a description following this format:

## What
One sentence explaining what this PR does.

## Why
Brief context on why this change is needed

## Changes
- Bullet points of specific changes made
- Group related changes together
- Mention any files deleted or renamed
```

- **`name`** identifies the skill.
- **`description`** is the **matching criteria** — how Claude decides to use it.
- Everything after the second `---` is the instructions Claude follows once the skill activates.

## 3. Test it

Claude Code loads skills **at startup**, so **restart your session** after creating one. Verify it appears in the available-skills list. Then make some changes on a branch and say *"write a PR description for my changes"* — Claude indicates it's using the `pr-description` skill, checks your diff, and writes the description in your template every time.

## How skill matching works

1. On startup, Claude Code scans **four locations** for skills but loads **only the name + description** (not the full content).
2. On a request, it **semantically matches** your message against those descriptions — e.g. "explain what this function does" matches a skill described as "explain code with visual diagrams" because the intent overlaps.
3. On a match, Claude **asks you to confirm** loading the skill (so you stay aware of what context is pulled in).
4. After you confirm, it reads the full `SKILL.md` and follows it.

## Skill priority (name conflicts)

If a cloned repo's skill shares a name with one of yours, precedence is:

1. **Enterprise** — managed settings, highest
2. **Personal** — `~/.claude/skills`
3. **Project** — repo's `.claude/skills`
4. **Plugins** — installed plugins, lowest

So orgs can enforce standards via enterprise skills while individuals still customize. **Use descriptive names** to avoid conflicts (`frontend-review` / `backend-review`, not just `review`).

## Updating & removing

- **Update:** edit the skill's `SKILL.md`.
- **Remove:** delete its directory.
- **Always restart Claude Code** after changes for them to take effect.
