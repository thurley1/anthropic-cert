# Configuration and Multi-File Skills

A basic skill needs just `name` + `description`, but a few advanced techniques make skills much more powerful: extra metadata fields, reliable descriptions, tool restrictions, and multi-file organization.

## Metadata fields (SKILL.md frontmatter)

Two required, the rest optional:

| Field | Req? | Notes |
|-------|------|-------|
| **`name`** | ✅ | lowercase letters/numbers/hyphens only, ≤ 64 chars, should match the directory name |
| **`description`** | ✅ | ≤ 1024 chars; **the matching field** — most important |
| **`allowed-tools`** | optional | restricts which tools Claude may use while the skill is active |
| **`model`** | optional | which Claude model to use for the skill |

## Writing effective descriptions

Be explicit. "Your job is to help with docs" tells Claude nothing — same as it would tell you nothing.

A good description answers **two questions**:
1. **What** does the skill do?
2. **When** should Claude use it?

If a skill isn't triggering when expected, **add keywords** that match how you actually phrase requests — the description is the matching signal, so the language matters.

## Restricting tools with `allowed-tools`

For read-only or security-sensitive workflows, limit what the skill can do:

```markdown
---
name: codebase-onboarding
description: Helps new developers understand how the system works.
allowed-tools: Read, Grep, Glob, Bash
model: sonnet
---
```

While this skill is active, Claude can use **only** those tools without asking — no editing/writing. Omit `allowed-tools` entirely and the skill doesn't restrict anything (normal permission model).

## Progressive disclosure (multi-file skills)

Activating a skill loads its `SKILL.md` into context. Cramming a 2,000-line file in has two costs: it eats context, and it's painful to maintain.

**Progressive disclosure:** keep essential instructions in `SKILL.md`, and put detailed material in separate files Claude reads **only when needed**. Suggested layout:

```
my-skill/
├── SKILL.md          # essentials + links to the below
├── scripts/          # executable code
├── references/       # extra documentation
└── assets/           # images, templates, data
```

In `SKILL.md`, link to supporting files with clear "load this when…" instructions — e.g. Claude reads `references/architecture-guide.md` only when someone asks about system design, not when they ask where to add a component. It's a **table of contents** in context, not the whole document.

**Rule of thumb: keep `SKILL.md` under ~500 lines**; beyond that, split into reference files.

## Using scripts efficiently

Scripts in the skill dir **run without loading their contents** into context — only the **output** consumes tokens. The key instruction in `SKILL.md`: tell Claude to **run** the script, not **read** it.

Great for:
- environment validation
- consistent data transformations
- operations more reliable as **tested code** than as generated code
