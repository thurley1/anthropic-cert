# Workflows vs Agents

The section's capstone decision. Both handle multi-step tasks; they differ in **who owns the control flow**.

- **Workflow** — a *predefined* series of Claude calls for a known problem. You picture the steps ahead of time and break the task into small, focused subtasks. **You** drive.
- **Agent** — Claude gets a goal + basic tools and *formulates its own plan*, combining tools in unexpected ways. **Claude** drives.

> Concept lesson — no code. (We built runnable versions of both this section: parallelization/chaining/routing/evaluator-optimizer are workflows; the datetime + file demos were agents.)

## Trade-offs

| | Workflows | Agents |
|---|-----------|--------|
| **Control flow** | you (code) | Claude (the loop) |
| **Accuracy** | higher — one focused subtask at a time | lower — juggles planning + execution |
| **Reliability / predictability** | high | lower, less predictable |
| **Test / evaluate** | easy — you know each step | hard — unknown step sequence |
| **Flexibility** | low — solves specific tasks | high — handles novel, varied requests |
| **UX** | constrained (known inputs) | flexible; can ask the user for more input |
| **Upfront work** | more planning/design | less; more runtime unpredictability |

## The rule of thumb

> **Your job is to solve problems reliably.** Users don't care that it's a "fancy agent" — they want something that works consistently.

**Default to workflows. Reach for agents only when truly required.**

- **Workflow** when you have a **well-defined process** and can enumerate the steps/inputs — most production features.
- **Agent** when requirements **can't be predetermined** — varied, unpredictable requests that need creative tool-combining or user back-and-forth.

And it's not binary: you can **compose** them — a workflow whose one step is an agent, an agent that internally runs a workflow, a router that dispatches to either. Start with the simplest tier that solves the problem (single call → workflow → agent), matching the guidance from the very first tool-use decision.

## Section wrap

Patterns covered:
- **Evaluator–Optimizer** — produce → grade → refine
- **Parallelization** — fan out independent subtasks → aggregate
- **Chaining** — sequential steps that build on each other
- **Routing** — categorize → dispatch to a specialized pipeline
- **Agents** — goal + abstract, combinable tools; Claude plans the steps
- **Environment inspection** — observe results (read-before-write, screenshots) to adapt

These are repeatable recipes — identify the shape of your problem, pick the pattern (or compose several), then write the code.
