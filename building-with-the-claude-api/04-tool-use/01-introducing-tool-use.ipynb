# Introducing Tool Use

By default Claude only knows its training data — no current events, no real-time data, no access to your systems. Ask *"What's the weather in San Francisco?"* and it can only say *"I don't have access to up-to-date weather information."*

**Tools** fix this by giving Claude a structured way to *request* information (or actions) that your code then fulfills.

## The tool-use loop

Tool use is a back-and-forth between **your app** and **Claude** — Claude never runs code itself; it *asks*, you *execute*, you *return*.

```
1. Initial request   You → Claude:  the question + the tools Claude may use
2. Tool request      Claude → You:  "call get_weather with {location: 'San Francisco'}"  (stop_reason: tool_use)
3. Data retrieval     You:          run the real code / hit the real API
4. Final response    You → Claude:  the tool result  →  Claude writes the final answer
```

Steps 2–4 can repeat: Claude may call several tools (or the same tool several times) before it has enough to answer. The loop ends when Claude responds with `stop_reason: end_turn` instead of `tool_use`.

## Weather example, concretely

| Step | What happens |
|------|--------------|
| 1 | User asks "What's the weather in San Francisco?"; your request also declares a `get_weather` tool |
| 2 | Claude decides it needs live data and emits a **tool_use** block requesting `get_weather(location="San Francisco")` |
| 3 | Your server calls a real weather API and gets, say, `62°F, foggy` |
| 4 | You send that back as a **tool_result**; Claude replies "It's currently 62°F and foggy in San Francisco." |

## Why it matters

- **Real-time information** — data that didn't exist at training time (weather, prices, your DB)
- **External system integration** — databases, APIs, internal services
- **Dynamic responses** — answers grounded in the latest data
- **Structured interaction** — Claude states *exactly* what it needs and how to ask for it

Tool use turns Claude from a static knowledge base into a dynamic assistant that works with live data — and it's the same mechanism that underlies agents (a loop of tool calls) later in the course.
