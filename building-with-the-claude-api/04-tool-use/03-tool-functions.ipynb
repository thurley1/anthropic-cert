# Tool Functions

A **tool function** is just a plain Python function that Claude can ask you to run when it needs extra data or an action. Ask *"What time is it?"* and Claude would call your date/time function to find out.

This lesson writes the **first** of the reminder project's three tools — with *no Claude involved yet*. The next lesson (**Tool schemas**) describes it to Claude; the one after wires it into the chat loop.

**Best practices for tool functions:**
- **Descriptive names** — function *and* parameter names should make the purpose obvious (Claude reads them).
- **Validate inputs** — check required params aren't empty/invalid and `raise` when they are.
- **Meaningful error messages** — Claude *sees* the error and may retry with corrected arguments. A clear `"Location cannot be empty"` lets it self-correct.

## Tool 1 — get the current date/time

A `date_format` parameter lets Claude request the time in whatever shape it needs. The default gives `year-month-day hour:minute:second`.

```python
from datetime import datetime


def get_current_datetime(date_format="%Y-%m-%d %H:%M:%S"):
    if not date_format:
        raise ValueError("date_format cannot be empty")
    return datetime.now().strftime(date_format)
```

## Test it with different formats

```python
# Default format, e.g. "2026-01-15 14:30:25"
get_current_datetime()
```

```python
# Just hour and minute, e.g. "14:30"
get_current_datetime("%H:%M")
```

## The validation in action

An empty format is unlikely, but the check demonstrates the pattern: raise a clear error that Claude could read and recover from.

```python
try:
    get_current_datetime("")
except ValueError as e:
    print(f"Raised as expected: {e}")
```
