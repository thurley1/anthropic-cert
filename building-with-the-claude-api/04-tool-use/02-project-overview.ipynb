# Project Overview — A Reminder System

The section's running project: teach Claude to **set reminders for future dates**. The goal is natural language like:

> "Set a reminder for my doctor's appointment. It's a week from Thursday." → *"OK, I will remind you."*

Simple on the surface, but it exposes three real gaps between what Claude can do unaided and what the task needs — and tools are how we bridge each gap.

> Concept lesson — no code yet. We implement the tools one at a time, starting next with **Tool functions**.

## Why it's harder than it looks

| Limitation | Problem |
|------------|---------|
| **Limited time awareness** | Claude may know the date but not the exact current time |
| **Date-calculation issues** | Claude isn't reliable at date arithmetic, especially many days out ("a week from Thursday") |
| **No reminder capability** | Claude has no built-in mechanism to actually *set* a reminder |

The guiding principle: when the model has a limitation, **extend it with a tool** rather than fighting the limitation in the prompt.

## The three tools

| Tool | Purpose | Bridges |
|------|---------|---------|
| **get current datetime** | Give Claude the precise current date & time | limited time awareness |
| **add duration to datetime** | Reliable date arithmetic (base time + duration → target time) | date-calculation issues |
| **set reminder** | Actually record a reminder in the system | no reminder capability |

By the end, Claude handles "remind me in a week" by **chaining** these: get *now* → add the duration → set the reminder at the computed time. We build them one at a time (simplest first) so the tool-calling mechanics are clear before the multi-tool orchestration.
