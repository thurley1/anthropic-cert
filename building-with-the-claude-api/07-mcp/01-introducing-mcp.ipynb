# Introducing MCP

**Model Context Protocol (MCP)** is a communication layer that gives Claude context and tools **without you writing the integration code**. It shifts the burden of tool definitions and execution off your server and onto specialized **MCP servers**.

> Concept lesson — no code. The build (server, then client) starts next at **Project setup**.

## The architecture

```
MCP Client  (your app)  ──connects to──▶  MCP Server(s)
                                           ├─ tools
                                           ├─ resources
                                           └─ prompts
```

Each **MCP server** is an interface to some outside service, packaging its functionality as reusable **tools**, **resources**, and **prompts**. Your app is the **MCP client** that connects to one or more servers and uses what they expose.

## Why it exists — the GitHub example

Say you're building a chat where users ask about their GitHub data: *"What open PRs are there across all my repositories?"* Claude needs tools to hit GitHub's API.

**Without MCP:** *you* author every tool — repos, PRs, issues, projects, ... — each needing a schema **and** a function implementation you write, test, and maintain. GitHub is huge, so that's a mountain of integration code.

**With MCP:** those tools already live inside a dedicated **GitHub MCP server**. You connect to it and use its pre-built tools — no implementation on your side.

## Common questions

- **Who authors MCP servers?** Anyone can. Often the service provider ships an official one (e.g. an AWS MCP server for their services).
- **How is it different from direct API calls?** A direct API call means *you* write the tool schemas/functions. An MCP server hands you those already defined.
- **Isn't this just tool use?** No — they're complementary. Tool use is the *mechanism*; MCP is about **who builds and maintains** the tools. With MCP, someone else already wrote the schemas and functions and packaged them in a server you connect to.

## What this section builds (and a modern note)

Over the next lessons we build **both sides**: an MCP **server** (exposing tools, then resources, then prompts) and an MCP **client** that connects to it — plus the **MCP Inspector** for debugging. This will use the `mcp` Python package and standalone `.py` files (not just notebook cells), so the structure shifts a bit from earlier sections.

> **Modern note:** MCP has evolved since the course. The two current **transports** are **stdio** (local subprocess — what this section uses) and **Streamable HTTP** (remote), the latter having superseded the older HTTP+SSE transport. The core concepts — tools / resources / prompts and the client↔server handshake — are unchanged; only some transport/plumbing details have moved.

> **Codebase aside:** MCP is exactly how a product like Code Navigator would expose codebase capabilities (search, symbol lookup, diff) to *any* MCP-speaking client — write the server once, and Claude Code, claude.ai, or your own app can all use it.
