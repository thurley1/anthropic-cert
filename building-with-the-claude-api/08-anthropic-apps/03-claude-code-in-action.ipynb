# Claude Code in Action

Claude Code is a **collaborative partner across the whole project** — setup to deploy — not just a code generator. The more context and structure you give it, the better it performs.

> Workflow/demo lesson. The sample project is in [`./app-starter/`](./app-starter) (a small MCP document-tools package) — open Claude Code *in that folder* to practice these workflows.

## `/init` and `CLAUDE.md`

Run **`/init`** first: Claude scans the codebase and writes a **`CLAUDE.md`** summarizing structure, dependencies, style, build commands, and patterns. That file is auto-included as context in every future conversation.

**Scopes (multiple `CLAUDE.md` files):**

| Scope | Shared? | Use |
|-------|---------|-----|
| **Project** | committed, all engineers | project rules/conventions |
| **Local** | not in git | your personal project notes |
| **User** | across all your projects | your global preferences |

Quick-add a rule with the **`#`** command: type `# Always use descriptive variable names` and Claude offers to save it to project / local / user memory.

## The core workflow: context → plan → implement

Claude works best as an **effort multiplier** when you structure the task:

1. **Feed context** — point Claude at the files relevant to the feature; have it read/analyze them so it learns your patterns.
2. **Plan first (no code yet)** — ask for an approach + steps, explicitly *not* an implementation. Review/refine the plan.
3. **Implement** — once the plan's solid, ask Claude to build it (and run tests, commit, etc.).

Separating planning from coding catches wrong approaches before they become wrong code.

## Test-driven variant

Often more robust because Claude gets clear success criteria:

1. Feed context (relevant files).
2. Ask Claude to **brainstorm test cases** for the feature.
3. Ask it to **implement the chosen tests**.
4. Ask it to **write code until the tests pass** — it iterates against the suite.


## Practical example (against `app-starter`)

Adding a `document_path_to_markdown` tool that builds on the existing `binary_document_to_markdown`:

```text
# 1. context
> Read the tools/math.py and tools/document.py files

# 2. plan (not implementation)
> Plan to implement a document_path_to_markdown tool:
  1. A function that: takes a file path, validates it exists, determines the
     file type from the extension, reads the binary data, and leverages the
     existing binary_document_to_markdown function; returns a markdown string
  2. Add documentation
  3. Register the tool with the MCP server (main.py)
  4. Add tests

# 3. implement
> Implement the plan
```

Claude writes the function, wires it into `main.py`, adds tests, and runs the suite to verify — the same read → plan → implement → test loop, end to end.

## Handy commands & automation

- **`/init`** — scan the codebase, write `CLAUDE.md`
- **`/clear`** — reset conversation history/context (do this between unrelated tasks to keep context clean)
- **`#`** — append a note/rule to a `CLAUDE.md`
- Routine dev work: staging/committing to git, running tests, managing deps — ask Claude instead of switching to the terminal yourself.

> **Meta / currency:** this entire certification repo is a live example — a project-scope `CLAUDE.md`-style workflow where Claude Code reads files, plans, implements, and commits across dozens of lessons. Modern Claude Code adds far more than the course shows (custom `/`-slash-commands, subagents, hooks, skills, plan mode, MCP), but `/init`, `/clear`, `#`, and `CLAUDE.md` are still the everyday core.
