# Introduction

Welcome to **Claude Code in Action**. The course teaches Claude Code — Anthropic's agentic coding assistant — from install through hands-on use, custom commands, MCP, GitHub integration, hooks, and the SDK.


## What this course covers

| Section | Topics |
|---------|--------|
| **What is Claude Code?** | the concept of a coding assistant; what Claude Code is; a look at it in action |
| **Getting hands on** | setup, project setup, adding/controlling context, making changes, custom commands, MCP, GitHub integration |
| **Hooks and the SDK** | lifecycle hooks and the Claude Code SDK for programmatic/automated use |
| **Wrapping up** | quiz + next steps |

The throughline: Claude Code is a real **agent** (goal + tools + a loop), and the course shows how to drive it effectively and extend it.

> Meta: this repo itself was built *with* Claude Code, so you've already seen the end state of everything this course teaches.
