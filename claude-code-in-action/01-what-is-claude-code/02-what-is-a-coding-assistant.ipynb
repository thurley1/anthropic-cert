# What Is a Coding Assistant?


## The idea

A **coding assistant** is an AI that helps you write and change software. They fall on a spectrum:

- **Autocomplete / inline suggestions** (e.g. editor tab-completion) — predicts the next lines as you type. Narrow scope, no project-wide action.
- **Chat assistants** — answer questions, draft snippets, but *you* copy/paste and wire things up.
- **Agentic assistants** (Claude Code) — you give a **goal**; the assistant **reads files, runs commands, edits code, runs tests**, and iterates until the task is done. It takes actions, not just suggestions.

## Where Claude Code sits

Claude Code is the **agentic** kind, and specifically **terminal-native and project-aware**:

- Works across your **whole project** (not one open file) — finds and reads what it needs.
- **Uses tools**: read/edit/write files, run bash, search, fetch the web, plus MCP servers.
- Runs a **loop**: act → observe results → adjust → repeat, until the goal is met.

It's the same request → tool-use → result → repeat loop you build against the Claude API directly — packaged as a polished, general-purpose coding agent.
