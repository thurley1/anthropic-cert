# Claude Code in Action

> **Synthesized recap — not verbatim course content.** The lesson source wasn't captured, so this is a current, accurate write-up of the topic; the course's exact wording/examples may differ.

A high-level look at what using Claude Code feels like, before the hands-on section.

## A typical session

1. **Open it in a project:** run `claude` in the repo. It picks up your `CLAUDE.md` (project context).
2. **Describe a goal** in plain English — "add a `/health` endpoint and a test for it."
3. **It explores** — greps/reads the relevant files to learn your patterns (inspect before acting).
4. **It plans** (optionally in Plan Mode) — proposes the approach; you can approve or steer.
5. **It implements** — edits files (showing diffs), writes tests, runs them via bash, and fixes failures.
6. **It wraps up** — can stage and commit to git, and summarize what changed.

Throughout, you stay in the loop: approve edits/commands per your permission mode, redirect, or `/clear` and restart.

## Why it's powerful

- **Whole-project reach** + real tools → it handles tasks a snippet generator can't (multi-file changes, "run the tests and fix what breaks").
- **Observes results** → catches and corrects its own mistakes (tests fail → it iterates).
- **Extensible** → custom commands, MCP servers, hooks, and the SDK tailor it to your workflow (the rest of the course).
