# Claude Code Setup

Claude Code is Anthropic's agentic coding assistant that runs in your terminal (and now IDEs, desktop, and web). This lesson: get it installed, authenticated, and running.


## Install

Two routes (pick one):

- **Native installer (no Node):**
  - macOS/Linux: `curl -fsSL https://claude.ai/install.sh | bash`
  - Windows (PowerShell): `irm https://claude.ai/install.ps1 | iex`
- **npm:** `npm install -g @anthropic-ai/claude-code` (needs Node.js — check with `npm help`)

Runs natively on macOS, Linux, and **Windows** (PowerShell) — WSL no longer required.

## Authenticate & run

1. In a project directory, run **`claude`**.
2. First launch prompts you to log in — two options:
   - **Anthropic account** (Claude Pro/Max subscription), or
   - **API key** from the Console (usage billed to the key).
3. You land in the interactive **REPL**. Type a request in plain English; Claude reads files, runs commands, and edits code to accomplish it.

Useful startup commands: `claude` (interactive), `claude "one-off prompt"` (headless-ish), `claude --help`, `/login` / `/logout`, `/status` (auth + model), `/model` (switch model).

## What you get out of the box

Built-in tools: **Read / Edit / Write** files, **Bash** (run commands), **Glob / Grep** (find/search), **WebFetch / WebSearch**, plus MCP support. It's an agent — you give a goal, it plans and uses these tools in a loop.
