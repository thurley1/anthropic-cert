# MCP Servers with Claude Code

Claude Code is an MCP **client**, so you extend it by connecting MCP **servers** (the payoff of the whole MCP topic).


## Registering a server

```bash
# local (stdio) server — note the -- before the launch command
claude mcp add <name> -- <command> [args...]
claude mcp add documents -- uv run main.py

# remote server
claude mcp add --transport http <name> <url>
claude mcp add --transport sse <name> <url>
```

**Scopes:** `--scope local` (default, just you) / `project` (writes a committed `.mcp.json`, shared with the team) / `user` (all your projects).

Manage them: `claude mcp list`, `claude mcp get <name>`, `claude mcp remove <name>`, and **`/mcp`** in-session to view connected servers + auth.

## Using what a server exposes

Once connected, its **tools** are available to Claude automatically, its **resources** can be referenced, and its **prompts** show up as slash-commands (typically `/mcp__<server>__<prompt>`).

**Popular servers:** Sentry (triage errors), Playwright (browser automation), Figma (designs), Atlassian (Jira/Confluence), Firecrawl (scraping), Slack (post/reply), GitHub. Combine several to match your workflow (read a Jira ticket → implement → run tests → post to Slack).

## Reminder: `mcp add` registers a server, it doesn't install a client

Claude Code *is* the client. `claude mcp add` just records how to reach a server (a command for stdio, a URL for remote). Build a server once → any MCP client (Claude Code, the Claude app, your own) can use it.
