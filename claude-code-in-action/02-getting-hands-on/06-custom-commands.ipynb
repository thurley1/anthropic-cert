# Custom Commands

> **Synthesized recap — not verbatim course content.** I didn't have this course's lesson text, so this is a current, accurate write-up of the topic in today's Claude Code (which has drifted from the course recording). Use it as a study reference; the exact course wording/examples may differ.

Turn a repeatable prompt into a reusable **slash-command** so you (and your team) can invoke it with `/name`.

## Defining one

A custom command is just a **Markdown file** whose body is the prompt:

- **Project:** `.claude/commands/<name>.md` → invoked as `/<name>` (committed, shared with the team)
- **User:** `~/.claude/commands/<name>.md` → available in all your projects

Example — `.claude/commands/review.md`:
```markdown
---
description: Review the current diff for bugs and style issues
argument-hint: [optional focus area]
---
Review the staged git diff. Focus on $ARGUMENTS.
Flag correctness bugs first, then style. Be concise.
```
Now `/review security` runs it with `$ARGUMENTS` = "security".

## Arguments & frontmatter

- **`$ARGUMENTS`** — everything after the command name.
- **`$1`, `$2`, ...** — positional arguments.
- **Frontmatter** (optional): `description` (shown in the `/` menu), `argument-hint`, `allowed-tools` (restrict what the command may do), `model`.
- Commands can run bash and reference files, so a command can be a whole mini-workflow (e.g. "run tests, summarize failures, propose fixes").

> Related, newer capability: **Skills** (`.claude/skills/`) package instructions + scripts/resources the model auto-invokes — a step beyond a single-prompt command. The course predates these; commands are the core primitive here.
