# Adding Context

Claude's output quality tracks the quality of the context you give it. This lesson: how to feed the right material into a task.


## Ways to add context

- **`@` file mentions** — type `@path/to/file` to pull a specific file into the prompt (with autocomplete). Point Claude at the files relevant to the task so it learns your patterns.
- **Let Claude find it** — you can also just describe what you need; Claude uses Glob/Grep/Read to locate and read files itself.
- **Paste** — paste text, error logs, or (in supported surfaces) **images/screenshots** directly.
- **`/add-dir`** — widen the working set to another directory.
- **URLs** — mention a URL and Claude can WebFetch it (docs, examples).

## The habit: context before action

Before asking Claude to build/fix something, have it **read the relevant files first**. Two benefits: it learns your existing conventions to build on, and it won't act blindly (the "environment inspection" principle from course 1 — read before you write).

More relevant context = better results; but don't dump the whole repo — irrelevant files are distractors and cost tokens. Be deliberate about *which* files matter for this task.
