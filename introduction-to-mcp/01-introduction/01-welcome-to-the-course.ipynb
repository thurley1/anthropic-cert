# Welcome to the Course

**Introduction to Model Context Protocol (MCP)** — course 3 of 4 in the CPN path.

MCP is an open standard for connecting AI applications to the tools, data, and context they need. Instead of every app hand-rolling a bespoke integration for every data source, MCP defines **one protocol** that any client (Claude Desktop, Claude Code, your own app) can speak to any server (a wrapper around a database, an API, a filesystem, etc.). Build the server once; every MCP-aware client can use it.

## What this course covers

| Section | Lessons |
|---------|---------|
| **Introduction** | Welcome · Introducing MCP · MCP clients |
| **Hands-on with MCP servers** | Project setup · Defining tools with MCP · The server inspector |
| **Connecting with MCP clients** | Implementing a client · Defining resources · Accessing resources · Defining prompts · Prompts in the client |
| **Assessment & wrap-up** | Final assessment · MCP review |

By the end you'll have built a working MCP **server** (exposing tools, resources, and prompts) and a **client** that connects to it — the two halves of the protocol.
