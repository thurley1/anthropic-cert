# Agents and Workflows

Both are strategies for tasks Claude **can't** finish in a single request. You've already built both this course — the tool-use loop where Claude decided what to do was an **agent**; the multi-step eval pipeline was a **workflow**.

> Concept lesson — no code. The coded patterns (parallelization, chaining, routing) start next.

## Workflows vs agents — how to choose

The deciding question: **how well do you know the task up front?**

| | Workflow | Agent |
|---|----------|-------|
| **What** | a predetermined series of calls to Claude | give Claude a goal + tools, let it figure out the steps |
| **Use when** | you can picture the exact steps, or your UX constrains users to set tasks | you don't know the exact task/parameters in advance |
| **Control** | you own the control flow (code) | Claude owns the control flow (the loop) |

Neither is "better" — workflows give predictability and cheaper/tighter control; agents give flexibility for open-ended tasks. Start with the simplest thing that fits (often a workflow).

## Example — image → CAD (a workflow)

A web app where a user drops an image of a metal part and you produce a **STEP** file (3D model). You know exactly what to do for a given image, and can write it as fixed steps → perfect **workflow**:

1. Feed the image to Claude → describe the object
2. From the description → ask Claude to model it with the **CadQuery** library
3. Create a rendering
4. Ask Claude to **grade** the rendering against the original image; if it's off, fix and repeat

## The Evaluator–Optimizer pattern

That grading loop is a reusable pattern:

```
        ┌──────────────┐        output        ┌──────────┐
input → │   Producer   │ ───────────────────▶ │  Grader  │
        └──────────────┘                      └────┬─────┘
               ▲            reject + feedback       │ accept
               └───────────────────────────────────┤
                                                     ▼
                                                 final output
```

- **Producer** — takes input, creates output (Claude modeling the part + rendering).
- **Grader** — evaluates output against criteria.
- **Feedback loop** — on reject, feedback returns to the producer.
- **Iteration** — repeat until the grader accepts.

This is exactly the **model-grader** idea from the *Prompt evaluation* section, wired into a *generation* loop instead of an offline eval — the grader becomes a runtime quality gate.

## Why learn the patterns

Identifying a workflow doesn't build anything by itself — you still write the code. But these patterns are **repeatable recipes** proven across many apps. This section covers a handful:

- **Parallelization** — fan out independent subtasks, gather results
- **Chaining** — pipe one step's output into the next
- **Routing** — classify the input, dispatch to the right handler
- **Evaluator–Optimizer** (this lesson) — produce → grade → refine

...then contrast all of them with true **agents**.
