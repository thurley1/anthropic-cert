# Anthropic Apps — Claude Code & Computer Use

This module looks at two Anthropic-built applications as **case studies in AI agents**: **Claude Code** and **Computer Use**. Understanding how they work sets up building your own agents in the next section.

> Concept lesson — no code. This section is mostly demos/walkthroughs, so these notebooks are recaps rather than runnable exercises.

## The plan

1. **Claude Code** — an agentic coding assistant that runs in your terminal (edit files, fix bugs, answer coding questions, drive dev workflows). We'll set it up and use it on a small project.
2. **Computer Use** — tools that let Claude interact with a full desktop: browse the web, use GUI apps, navigate visual interfaces. Far beyond text-only.
3. **Agents** — what makes these successful *as agents*, informing your own agent work.

## Why these matter for agents

Both are real, shipping agents that demonstrate the core principles:

- **Tool integration & usage** — the model acts through tools (files, bash, browser, screen).
- **Multi-step task execution** — a goal decomposed into many tool calls.
- **Environmental interaction** — reading state (files, screenshots) and changing it.
- **Autonomous problem-solving** — looping, observing results, self-correcting until done.

These are exactly the ingredients we'll formalize in **Agents and workflows** — Claude Code and Computer Use are just the polished, real-world versions of the loop you built by hand in the tool-use section.
