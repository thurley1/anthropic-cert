# Enhancements with MCP Servers

Claude Code has an **MCP client built in**, so you extend it by connecting **MCP servers** — the exact payoff of the MCP section. Instead of being limited to Claude Code's built-in tools, you bolt on custom or third-party capabilities.

> Recap lesson — no code. This is the client side of what you built: you wrote a server; now register it with a client (Claude Code).

## What a server can add

Each MCP server exposes the three primitives from the MCP section:

- **Tools** — take actions
- **Prompts** — reusable templates
- **Resources** — access data

Claude Code surfaces all of them once connected.

## Registering a server

```bash
claude mcp add [server-name] [command-to-start-server]
```

Course example — a document server started by `uv run main.py`:

```bash
claude mcp add documents uv run main.py
```

Claude Code auto-connects to it on startup.

> **Currency notes:**
> - The modern CLI prefers a **`--` separator** before the launch command: `claude mcp add documents -- uv run main.py`. (Without it, flags after the name can be misparsed.)
> - **Scope:** `--scope local` (default, just you) / `project` (committed `.mcp.json`, shared) / `user` (all your projects).
> - **Remote servers:** `claude mcp add --transport http <name> <url>` (or `sse`) — connect to a hosted server instead of launching a local subprocess.
> - `claude mcp add` **registers** a server; it doesn't "install a client" (Claude Code *is* the client).

## Example — document processing

Register a server exposing `document_path_to_markdown` (the tool built in the last lesson on `app-starter`), then just ask:

> Convert the `tests/fixtures/mcp_docs.docx` file to markdown

Claude Code picks the custom tool automatically and returns the converted content — no manual wiring per request.

## Popular MCP integrations

The ecosystem has servers for common dev tools:

| Server | Adds |
|--------|------|
| `sentry-mcp` | discover + fix bugs logged in Sentry |
| `playwright-mcp` | browser automation for testing/troubleshooting |
| `figma-context-mcp` | expose Figma designs to Claude |
| `mcp-atlassian` | Confluence + Jira access |
| `firecrawl-mcp-server` | web scraping |
| `slack-mcp` | post/reply to Slack threads |

(Official first-party connectors for many of these also exist in the Claude app.)

## Building your workflow

The power is **combining servers** to match your process — e.g.:

- **Sentry** → fetch production error details
- **Jira** → read ticket requirements
- **Slack** → notify the team when work's done
- **Custom servers** → your internal tools/APIs

Now Claude Code works across the tools you already use — an assistant tailored to your stack.

> **Tie-back:** this is exactly the case we discussed for **StormBoard** — expose its capabilities as an MCP server, `claude mcp add` it, and Claude Code (and any MCP client) can drive analyses/rollups/codebase-Q&A without bespoke integration. Build the server once; every client benefits.
