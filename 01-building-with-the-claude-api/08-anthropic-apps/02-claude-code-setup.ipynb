# Claude Code Setup

Claude Code is a terminal-based coding assistant — Claude in your command line, with tools to work on real projects.

> Setup/walkthrough lesson — nothing to run in a notebook. (And you're already using it: this repo was built with Claude Code.)

## What it can do

- **File operations** — search, read, edit files in your project
- **Terminal access** — run commands from the conversation
- **Web access** — search docs, fetch examples
- **MCP server support** — add tools by connecting MCP servers (`claude mcp add …`) — the payoff of the last section

Runs on macOS, Linux, and Windows.

## Installation (course steps)

1. **Install Node.js** from nodejs.org (check with `npm help`).
2. **Install Claude Code:** `npm install -g @anthropic-ai/claude-code`
3. **Start & log in:** run `claude` in your terminal → it prompts you to log in to your Anthropic account.

Full guide: docs at Anthropic.

## Currency notes (things that changed since the course)

- **Native installer (no Node needed):** besides npm, there's a standalone installer — macOS/Linux `curl -fsSL https://claude.ai/install.sh | bash`, Windows PowerShell `irm https://claude.ai/install.ps1 | iex`. The `npm install -g @anthropic-ai/claude-code` route still works.
- **Native Windows:** the course lists "Windows WSL," but Claude Code now runs **natively on Windows** (PowerShell), not just under WSL — which is how this repo was built.
- **Login options:** you can authenticate with an **Anthropic account** (Claude Pro/Max subscription) *or* an **API key** from the Console — either works; the account route is what the course shows.
- **Way more than the course covers now:** subagents, hooks, skills, custom slash-commands, IDE extensions (VS Code/JetBrains), a web/desktop app, and MCP. The terminal REPL + `claude` command are still the core.
