# Introducing Retrieval Augmented Generation (RAG)

**RAG** lets you work with documents too big (or too many) to stuff into a single prompt. Instead of sending everything, you **break documents into chunks** up front, then at question time **retrieve only the most relevant chunks** and include those.

> Concept lesson — no code yet. Code starts next with **Text chunking strategies**.

## The problem

You have an 800-page financial doc and want to ask *"What risk factors does this company have?"* You have to get the relevant text to Claude somehow.

**Option 1 — stuff everything into the prompt:**

```
Answer the user's question about the financial document.

<user_question>
{user_question}
</user_question>

<financial_document>
{financial_document}
</financial_document>
```

Limitations: hard prompt-length limit, Claude gets less effective on very long prompts, and bigger prompts cost more and run slower.

**Option 2 — RAG:** chunk the document in preprocessing; at query time, find the chunks most relevant to the question (e.g. the "Risk Factors" section) and include only those.

## Trade-offs

| Benefits | Challenges |
|----------|------------|
| Claude focuses on only the relevant content | Requires a preprocessing/chunking step |
| Scales to very large documents | Need a search mechanism to find "relevant" chunks |
| Works across multiple documents | Retrieved chunks might miss needed context |
| Smaller prompts → cheaper + faster | Many ways to chunk — which is best? |

RAG trades **simplicity for scalability and efficiency**. Use it for very large documents, multiple documents, or when cost/latency/focus matter.

## Modern note — big context windows don't retire RAG

The course frames RAG mainly around the *"document won't fit"* problem. Current Claude models have **1M-token context windows**, so raw fit is far less often the blocker. But RAG still matters:

- **Cost & latency** — you pay for (and wait on) every input token; sending 3 relevant chunks beats sending 800 pages.
- **Focus / accuracy** — models still do better with tightly relevant context than with a haystack; irrelevant text is a distractor.
- **Scale beyond one window** — corpora (whole codebases, document collections) exceed even 1M tokens.
- **Freshness** — retrieval pulls current data the model never trained on.

So the *reason* shifts from "it doesn't fit" to "it's cheaper, faster, and more accurate to retrieve" — but the technique stays central.

## Aside — RAG over a codebase

"Chat with a repository" (à la a code-navigation assistant) is a RAG problem: chunk the source, index it, retrieve the pieces relevant to a question, and hand them to Claude. The decisions this section covers map directly onto that:

- **Chunking** — split by file / function / class rather than fixed character windows, so a chunk is a coherent unit of code.
- **Hybrid retrieval** — pair semantic embeddings with **BM25** lexical search (covered later): embeddings capture "what the code does," BM25 nails exact identifiers/symbols that embeddings often blur.
- **Retrieval is evaluable** — the eval harness from the prompt-evaluation section applies: does the retrieved context actually contain the answer?
