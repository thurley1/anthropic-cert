# MCP Review

With the server built, here's the payoff: the three core **server primitives** and when to use each. The key insight — **each primitive is controlled by a different part of your stack.**

| Primitive | Controlled by | Serves |
|-----------|---------------|--------|
| **Tools** | the **model** (Claude) | give Claude new capabilities |
| **Resources** | your **app** code | get data into your app (UI / context) |
| **Prompts** | the **user** | predefined workflows triggered on demand |

## Tools — model-controlled

Claude decides when to call them, and uses the results directly to accomplish a task. Best for giving Claude capabilities it can use **autonomously**.

Example: ask Claude to *"calculate the square root of 3 using JavaScript"* — **Claude** decides to invoke a JS execution tool and run it.

## Resources — app-controlled

**Your application** decides when to fetch resource data and how to use it — usually for UI elements or to add context to a conversation.

In our project, resources were used two ways:

- Fetch data to populate **autocomplete** options in the UI.
- Retrieve content to **augment prompts** with extra context.

Real-world analogy: Claude's **"Add from Google Drive"** — the app decides which documents to show and injects their content into the chat context.

## Prompts — user-controlled

Triggered by **user** actions — button clicks, menu selections, slash commands. Ideal for on-demand, predefined workflows.

Real-world analogy: the **workflow buttons** below Claude's chat input — optimized workflows a user starts with a single click (our `/format` command is the same idea).

## Choosing the right primitive

- Need to give Claude **new capabilities**? → **Tool**
- Need to get **data into your app** for UI or context? → **Resource**
- Want **predefined workflows for users**? → **Prompt**

All three appear in Claude's own interface: workflow buttons = prompts, Google Drive integration = resources, code execution / calculations = tools.

**One line to remember:** tools serve the **model**, resources serve your **app**, prompts serve your **users**.
