## `list_prompts()`

Straightforward — call the session's list-prompts and return them:

```python
async def list_prompts(self) -> list[types.Prompt]:
    result = await self.session().list_prompts()
    return result.prompts
```

## `get_prompt()`

More interesting — it handles **variable interpolation**. The `args` you pass become keyword arguments to the server's prompt function:

```python
async def get_prompt(self, prompt_name, args: dict[str, str]):
    result = await self.session().get_prompt(prompt_name, args)
    return result.messages
```

If the server has a `format_document` prompt expecting `doc_id`, then `args = {"doc_id": "plan.md"}` gets interpolated into the template, and `result.messages` is the ready-to-send message list.

## How prompts work (recap)

A prompt defines a set of user/assistant messages clients can use. Keep them high-quality, well-tested, and relevant to the server's purpose. The workflow:

1. Write and evaluate a prompt for your server's functionality.
2. Define it with the **`@mcp.prompt`** decorator.
3. Clients can request it at any time.
4. Client-provided **arguments** become keyword arguments in your prompt function.
5. The function returns **formatted messages** ready for the model.

Reusable, parameterized prompts that stay consistent while allowing per-call customization through variables — especially useful for complex workflows where the model must receive properly structured instructions every time.
