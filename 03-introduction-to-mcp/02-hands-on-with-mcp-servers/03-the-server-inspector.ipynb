# The Server Inspector

You need a way to test an MCP server **without** wiring up a full application. The Python MCP SDK ships a built-in **browser-based inspector** for debugging and testing a server in real time.

## Starting the inspector

Activate your project environment (see the project README for the exact command), then run:

```bash
mcp dev mcp_server.py
```

With `uv`: `uv run mcp dev mcp_server.py` (from inside `cli-project/`).

It starts a dev server and prints a local URL, typically **`http://127.0.0.1:6274`**. Open it in your browser to reach the MCP Inspector.

> Adaptation: `mcp dev` launches the inspector via **`npx @modelcontextprotocol/inspector`**, so **Node.js/`npx` must be installed** (you already have Node). The command downloads the inspector on first run. Port is `6274` by default.

## The inspector interface

The UI is under active development, so it may look different, but the core pieces are consistent:

- A **Connect** button to start your MCP server.
- Navigation tabs for **Resources, Tools, Prompts**, and more.
- A tools listing + testing panel.

Click **Connect** first to initialize the server — the status flips from *Disconnected* to *Connected*.

## Testing your tools

Open the **Tools** section and click **List Tools** to see everything your server exposes. Select a tool and the right panel shows its details and input fields.

To test the document reader:

1. Select **`read_doc_contents`**.
2. Enter a document id (e.g. `deposition.md`).
3. Click **Run Tool**.
4. Check the result for success and the expected output.

The inspector shows both the **success status** and the **returned data**, so you can verify the tool directly.

## Testing tool interactions

Run tools **in sequence** to verify workflows. For example: use `edit_document` to change a document, then immediately run `read_doc_contents` to confirm the edit applied.

The inspector **keeps server state between calls**, so edits persist across runs and you can check end-to-end behavior.

## Why it matters (dev workflow)

The inspector replaces writing throwaway test scripts or standing up a full app. In one loop you can:

- Iterate quickly on tool implementations.
- Test edge cases and error conditions (e.g. a bad `doc_id` raising `ValueError`).
- Verify tool interactions and state.
- Debug in real time.

That immediate feedback catches issues early and makes server development much faster.
