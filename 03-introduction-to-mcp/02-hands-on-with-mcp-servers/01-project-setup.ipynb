## Using the CLI

- **Chat:** type a message and press Enter.
- **Document retrieval:** `@` + a doc id pulls that document into your query тАФ `> Tell me about @deposition.md`.
- **Commands:** `/` runs a prompt defined by the server тАФ `> /summarize deposition.md` (Tab auto-completes).

`@` mentions rely on **resources**; `/` commands rely on **prompts** тАФ both are TODOs you'll implement later in the course.
