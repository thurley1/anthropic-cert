# Introducing MCP

Model Context Protocol (MCP) is a communication layer that gives Claude context and tools without you writing tedious integration code. It shifts the burden of **tool definitions and execution** off your server and onto specialized MCP servers.

## Architecture at a glance

The basic picture: an **MCP Client** (living inside your application) connects to one or more **MCP Servers**, each exposing **tools, prompts, and resources**. Every MCP server acts as an interface to some outside service.

```
Your app  ──►  MCP Client  ──►  MCP Server(s)  ──►  outside service (GitHub, AWS, DB, ...)
                                 tools / prompts / resources
```

> Terminology note: the course calls the client side "your server." In practice the **client is embedded in whatever you're building** (a chat backend, Claude Desktop, Claude Code). It's the piece that speaks the protocol out to servers.

## The problem MCP solves

Say you're building a chat interface where users ask Claude about their **GitHub** data, e.g. *"What open pull requests are there across all my repositories?"* Claude needs tools to reach GitHub's API.

GitHub is huge: repositories, pull requests, issues, projects, and much more. **Without MCP**, you'd hand-author a tool schema + function for every feature you want to cover, then write, test, and maintain all of it. That's a large and ongoing burden.

## How MCP works

MCP moves tool definitions and execution to a **dedicated MCP server**. Instead of you authoring all those GitHub tools, an **MCP server for GitHub** wraps that functionality and exposes it as a standardized set of tools. Your application just **connects** to that server rather than implementing everything from scratch.

## MCP servers explained

An MCP server provides access to data or functionality from an outside service, exposing **tools, prompts, and resources** in a standardized way.

In the GitHub example, the server offers tools like `get_repos()` and connects directly to GitHub's API. Your app talks to the MCP server; the server handles all the GitHub-specific details.

## Common questions

**Who authors MCP servers?**
Anyone can. Often the service provider ships an official one (e.g. AWS releasing an official MCP server with tools for their services).

**How is this different from calling APIs directly?**
An MCP server gives you tool schemas and functions **already defined**. Call an API directly and you author those tool definitions yourself. MCP saves that implementation work.

**Isn't MCP just tool use?**
Common misconception. They're **complementary, not the same**. An MCP server *provides* the tool schemas and functions; **tool use** is *how Claude actually calls* tools. The difference is **who does the work**: with MCP, someone has already implemented the tools for you.

## Takeaway

Instead of building and maintaining a complex set of integrations yourself, you connect to MCP servers that do the heavy lifting of reaching external services. Build (or adopt) the server once; any MCP-aware client can use it.
