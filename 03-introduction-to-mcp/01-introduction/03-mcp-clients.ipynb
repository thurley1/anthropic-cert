# MCP Clients

The MCP **client** is the communication bridge between your application and MCP servers. It's your access point to the tools a server provides, and it handles the message exchange and protocol details so your app doesn't have to.

## Transport-agnostic communication

A key strength of MCP: it's **transport agnostic**. The client and server can talk over different protocols depending on your setup.

- **Most common:** client and server run on the **same machine**, communicating over **standard input/output (stdio)**.
- Also supported: **HTTP**, **WebSockets**, and other network protocols.

Same protocol, same message types, regardless of the transport underneath.

## MCP message types

Once connected, client and server exchange message types defined in the MCP spec. The two you'll work with most:

| Request → Result | What it does |
|------------------|--------------|
| **`ListToolsRequest` → `ListToolsResult`** | Client asks *"what tools do you provide?"* and gets back the list of available tools. |
| **`CallToolRequest` → `CallToolResult`** | Client asks the server to run a specific tool with given arguments, and receives the results. |

## How it all works together

A user asks *"What repositories do I have?"* Here's the full round-trip, from your app through the client to GitHub and back to Claude:

1. **User query** — the user submits the question to your app.
2. **Tool discovery** — your app needs to know what tools exist to send to Claude.
3. **List-tools exchange** — your app asks the MCP client for available tools.
4. **MCP communication** — the client sends a `ListToolsRequest` to the server and gets a `ListToolsResult`.
5. **Claude request** — your app sends the user's query **plus** the available tools to Claude.
6. **Tool-use decision** — Claude decides it needs a tool to answer.
7. **Tool-execution request** — your app asks the MCP client to run the tool Claude specified.
8. **External API call** — the client sends a `CallToolRequest` to the server, which makes the actual GitHub API call.
9. **Results flow back** — GitHub responds; data returns through the server as a `CallToolResult`.
10. **Tool result to Claude** — your app sends the tool results back to Claude.
11. **Final response** — Claude formulates an answer using the repository data.
12. **User gets answer** — your app delivers Claude's response to the user.

## Why the many steps are fine

Each component has one clear responsibility. The MCP client **abstracts away the server-communication complexity**, so you focus on application logic while still getting external tools and data.

Note the division of labor: **Claude** decides *whether/which* tool to call (steps 6); the **MCP client/server** handles *discovery and execution* (steps 4, 8-9); **your app** orchestrates the loop between them (steps 2, 5, 7, 10). That's the same tool-use loop from course 1, with the client/server pair standing in for hand-written tool functions.
